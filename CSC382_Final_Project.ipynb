```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Create an empty undirected graph
G = nx.Graph()

# Add 20 vertices labeled from 1 to 20
G.add_nodes_from(range(1, 21))

# Randomly create 20 unique edges to ensure our graph has at least 20 edges
edges = set()
while len(edges) < 20:
    v1, v2 = np.random.choice(range(1, 21), 2, replace=False)
    if v1 != v2:
        edges.add((v1, v2))

# Add the edges to the graph
G.add_edges_from(list(edges))

# Visualize the graph using matplotlib to see the nodes and edges
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_color='skyblue')
plt.title("Visual Representation of the Graph")
plt.show()

# Display the adjacency list to see direct connections of each vertex
print("Adjacency List:")
for line in nx.generate_adjlist(G):
    print(line)

# Display the adjacency matrix to see a matrix representation of connections
print("\nAdjacency Matrix:")
print(nx.adjacency_matrix(G).todense())

# Implement Depth-First Search (DFS) to explore the graph
# Starting from vertex 1 as the root node
dfs_tree = nx.dfs_tree(G, source=1)
print("\nDFS from vertex 1:")
print(list(nx.dfs_preorder_nodes(G, source=1)))

# Implement Breadth-First Search (BFS) to explore the graph
# This algorithm will also start from vertex 1 and explore layer by layer
bfs_tree = nx.bfs_tree(G, source=1)
print("\nBFS from vertex 1:")
print(list(nx.bfs_tree(G, source=1).edges()))

# Implement Dijkstra's Algorithm to find shortest paths from a single source
# Here, vertex 1 is the source, and we find shortest paths to all other vertices
path_lengths = nx.single_source_dijkstra_path_length(G, source=1)
print("\nShortest paths from vertex 1:")
for target, length in path_lengths.items():
    print(f"Vertex 1 to {target} has path length {length}")

# Visualize DFS and BFS trees to show how these algorithms traverse the graph
plt.figure(figsize=(15, 6))
plt.subplot(121)
nx.draw(dfs_tree, with_labels=True, node_color='lightgreen')
plt.title("DFS Tree from Vertex 1")

plt.subplot(122)
nx.draw(bfs_tree, with_labels=True, node_color='lightblue')
plt.title("BFS Tree from Vertex 1")
plt.show()
```

I implemented three graph traversal algorithms: BFS, DFS, and Dijkstra's algorithm. All corporal algorithms perform separate roles and have different input and output parameters.

Breadth First Search (BFS) is initiated at a given node and followed by all the nodes at the current vertex before moving to the nodes at the next level. The input is a graph and a starting node, while the output is the order of nodes visited. This highlights a level-order traversal of a graph, which is good for unweighted graphs where distances are the same.

On the other hand, depth-first search (DFS) works deeply in one path before going back to the previous path, so it is good for activities using paths, such as mazes. It is the same as BFS; however, the output tends to display a conceptually less apparent direction that optimizes the depth traversal compared to backtracking.

Dijkstra's Algorithm solves each shortest-path problem in that it deals with graphs with non-negative edge weights only. It is used to estimate the shortest form of connectivity from one point or node to another in weightage networks such as maps for navigation.

The Edge List is particularly applicable for cases where one needs to access/inspect/move around with the exact edges that are being listed, for example when implementing algorithms that alter edge weights, or in subgraph generation. This approach is easy and works well if the emphasis is on the connections (springs) rather than on the characteristics of the vertices. I found out that if the application or analysis is more concerned with the edges rather than the nodes, the Edge List can be very easy to manage.

The effective use of such algorithms also varies depending on the particular need that the application will serve. BFS is well suited when all edges are uniform whether in terms of distance or cost since all edges are equal for BFS when the graph is unweighted. DFS is used for looks for all means, especially when a search algorithm must reach every vertex as in games or any other applications where a full path assessment is required. The far reaching application of Dijkstra's Algorithm is in the weighted graphs used in GPS and Networks to determine the minimum distance learn in travel distance or time. These algorithms were run on a graph with 20 nodes and 20 edges, which captured the capabilities of each and when the graph and edge characteristics are best applied to each.
